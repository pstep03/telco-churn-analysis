```python
#Import libraries for data manipulatuion and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Display all columns when inspecting the dataframe
pd.set_option("display.max_columns", None)
```

```python
#Load the raw Telco customer dataset
df = pd.read_csv("raw_telco_data.csv")

#Preview the dataset to understand structure and variables
df.head()
```

```python
#Check data types and identify potential missing values
df.shape
df.info()
```

```python
#Summary statistics for numerical variables
df.describe()
```

```python
#Check how many missing values
df.isnull().sum()
```

```python
#Check for duplicates
df.duplicated().sum()
```

```python
#Convert TotalCharges to numeric
#Some values may be stored as strings due to improper or empty entries
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
#Remove rows where TotalCharges is missing
#Value is required for modeling
df = df.dropna(subset=["TotalCharges"])
```

```python
#Create tenure groups to categorize customer length of service
df["tenure_group"] = pd.cut(df["tenure"], bins=[0,12,24,48,72])
```

```python
#Convert Churn from Yes/No to binary values for machine learning models
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
```

```python
#Final checks to confirm dataset structure after cleaning
df["Churn"].dtype
df["Churn"].value_counts()
```

```python
#Final checks to confirm dataset structure after cleaning
df.shape
df.info()
```

```python
#Save cleaned dataset for use in modeling notebooks
df.to_csv("clean_telco_data.csv", index=False)
```
